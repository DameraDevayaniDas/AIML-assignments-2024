```python
num1 = int(input("Enter First Number: "))
num2 = int(input("Enter Second Number: "))

print("Enter which operation would you like to perform?")
ch = input("Enter any of these char for specific operation +,-,*,/: ")

result = 0
if ch == '+':
    result = num1 + num2
elif ch == '-':
    result = num1 - num2
elif ch == '*':
    result = num1 * num2
elif ch == '/':
    result = num1 / num2
else:
    print("Input character is not recognized!")

print(num1, ch , num2, ":", result)
```

PART 2:visualisation of data using matplotlib,pylots packages

```python
import matplotlib.pyplot as plt
data_values = [1000, 2000, 2000, 3000, 3000, 3000, 4000, 4000, 5000, 5000, 5000, 5000, 6000, 6000, 7000]
plt.hist(data_values, bins=range(min(data_values), max(data_values) + 1), edgecolor='blue')
plt.xlabel('years of work')
plt.ylabel('income in rupees')
plt.title('my liftime income')
plt.show()
```

```python
import matplotlib.pyplot as plt
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [150, 300, 250, 400]
plt.bar(categories, values, color='blue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Chart')
plt.show()
```

```python
import matplotlib.pyplot as plt
x_values = [100, 200, 300, 400, 500]
y_values = [101, 120, 50, 80, 150]
plt.scatter(x_values, y_values, color='red', marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Scatter Chart')
plt.show()
```
